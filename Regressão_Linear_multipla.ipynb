```python
import numpy as np
```

```python
y_vetor = [220,260,320]
x = np.array([[1, 50, 5],[1, 60, 6],[1, 80, 7]])

y = np.array(y_vetor).reshape(-1,1)
xt = np.transpose(x)
mult = xt@x
inv = np.linalg.inv(mult)
betas = inv@xt@y
```
