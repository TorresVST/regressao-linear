```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = [20,22,25,27,30,32,35]
y = [45,50,60,65,75,80,90]

plt.scatter(x,y)
```

```python
modelo = LinearRegression()
modelo.fit(pd.DataFrame(x), y)
```

```python
a = modelo.coef_
l = modelo.intercept_ ## a: angular, l: linear
```

```python
print('Coeficiente angular: {0}\nCoeficiente linear: {1}'.format(a,l))
```

```python
plt.scatter(x,y)
plt.plot(x, l+a*x, color= 'red')
plt.show()
```
